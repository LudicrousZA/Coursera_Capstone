#### This is the Neighborhood Clustering Assignment Notebook.

```python
import numpy as np 
import pandas as pd 
import json
import matplotlib.cm as cm
import matplotlib.colors as colors
import folium 
import requests 

from pandas.io.json import json_normalize
from sklearn.cluster import KMeans
from geopy.geocoders import Nominatim 
from bs4 import BeautifulSoup

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

print('Libraries imported.')
```

#### Import Toronto Neighborhood Data and create BeautifulSoup Object

```python
data = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(data,'lxml')

from IPython.display import display_html
html_tab = str(soup.table)

# Display tab as 
display_html(html_tab,raw=True)
```

### Scrape table for Postal Code, Borough and Neighborhood values

```python
table_contents=[]
table=soup.find('table')
for row in table.findAll('td'):
    cell = {}
    if row.span.text=='Not assigned':
        pass
    else:
        cell['PostalCode'] = row.p.text[:3]
        cell['Borough'] = (row.span.text).split('(')[0]
        cell['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)
```

```python
df = pd.DataFrame(table_contents)
df.head()
```

### Display Dataframe shape

```python
df.shape
```

### Read Geospatial data into dataframe

```python
spatial_info = pd.read_csv('https://cocl.us/Geospatial_data')
spatial_info.head()
```

```python
df.keys()
```

### Correlate keys between geospatial and neighborhood dataframe for merging

```python
spatial_info.rename(columns={'Postal Code':'PostalCode'},inplace=True)
spatial_info.keys()
```

#### Merge latitude and longitude with original dataframe along PostalCode column

```python
df_ll = pd.merge(df,spatial_info,on='PostalCode')
df_ll.head()
```

### Create dataframe with Toronto Boroughs

```python
toronto_df = df_ll[df_ll['Borough'].str.contains('Toronto',regex=False)]
toronto_df.head()
```

### Create Toronto Map

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(toronto_df['Latitude'],toronto_df['Longitude'],toronto_df['Borough'],toronto_df['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
```

### Clustering of neighborhoods

```python
k=5

toronto_cluster = toronto_df.drop(['PostalCode','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_cluster)
kmeans.labels_

toronto_df.insert(0, 'Cluster Labels', kmeans.labels_)
```

### Map clustered neighborhoods in Toronto

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=12)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(toronto_df['Latitude'], toronto_df['Longitude'], toronto_df['Neighborhood'], toronto_df['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
